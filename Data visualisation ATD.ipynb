```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# do not use the first column as the index, set header to first row, skip spaces after delimiter
df_train = pd.read_csv('train.csv', index_col = False, header = 0, skipinitialspace = True)

# print head
df_train.head(5)
```

```python
# count occurence of specific labels:
df = df_train.labels.value_counts()
display(df)
```

```python
# create pie chart indicating the balance between classes
plot = df.plot.pie(figsize=(20, 20))
```

```python
# create bar plot for numerical indications
ax = df.plot.bar(x = 'labels', rot = 80)
```

```python
# create barplot: use sns seaplot for different colors for every class
plt.figure(figsize=(15,10))
labels = sns.barplot(x = df.index, y = df)
plt.xticks(rotation = 80)
```
